```python
def get_long_name(aminoPosamino):
    first_amino = aminoPosamino[0]
    last_amino = aminoPosamino[len(aminoPosamino)-1]
    dict = {'A': 'Ala', 'G': 'Gly', 'L':'Leu', 'M':'Met', 'F':'Phe',
           'W':'Trp', 'K':'Lys', 'Q':'Gln', 'E':'Glu', 'S':'Ser',
           'P':'Pro', 'V':'Val', 'I':'Ile', 'C':'Cys', 'Y':'Tyr', 'H':'His', 'R': 'Arg',
           'N':'Asn', 'D': 'Asp', 'T':'Thr'}
    
    aminoPosamino = aminoPosamino.replace(first_amino, dict.get(first_amino))
    aminoPosamino = aminoPosamino.replace(last_amino, dict.get(last_amino))
    return "p."+aminoPosamino
```

```python
naglu_data.with_column("Long Names", naglu_data.apply(get_long_name, 0)).num_rows
```
