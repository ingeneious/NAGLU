```python
def find_modified_codons(codons):
    for i in np.arange(3):
        if (str.istitle(codons[i])):
            ref = codons[i]
            break
    for j in np.arange(4, 7):
         if (str.istitle(codons[j])):
            alt = codons[j]
            break
    return make_array(ref, alt)
```

```python
#Testing table_joiner 
table2 = Table().with_columns("a2", [1, 1, 2],
                             "b2", [2, 2, 5],
                             "c2", [3, 3, 4],)
table2
```

```python
def table_joiner(table1, table2, name_for_table):
    array = make_array()
    for r in np.arange(table2.num_rows):
        if(table1.column(0).item(r) == table2.column(0).item(r)):
            if(table1.column(1).item(r) == table2.column(1).item(r)):
                if(table1.column(2).item(r) == table2.column(2).item(r)):
                    array = np.append(array, table1.column("d").item(r))
                else: #3rd element doesn't match
                     array = np.append(array, "")
            else:
                 array = np.append(array, "")
        else:
            array = np.append(array, "")
    return table2.with_column(name_for_table, array)
            
                    
                    
                
```

```python
#Testing table_joiner 
table_joiner(table1, table2, "extra stuff")
```

```python
def position_remover(pos):
    return int(pos[12:])

def location_and_codon_joiner_for_ouput(row):
    return str(row.item(0)) + find_modified_codons(row.item(12))[0] + find_modified_codons(row.item(12))[1]

def location_and_codon_joiner_for_answer(row):
    return str(row.item(0)) + row.item("Reference_Ans") + row.item("Alternate_Ans")


def output_table_modifier(output_tbl):
    """This method is for an output/results table. 
    It will take the location and ref/alt codons and combine them into one.
    It depends on find_modified_codons method and position_remover method."""
    output_tbl = output_tbl.with_column("Position", 
                           output_tbl.apply(position_remover, "Location"))
    output_tbl = output_tbl.move_to_start("Position")

    joined = output_tbl.with_column("Location+Codons", 
                                    output_tbl.apply(location_and_codon_joiner_for_ouput))
    joined = joined.move_to_start("Location+Codons")
    return joined

def input_table_modifier(input_tbl):
    """This method is for a CAGI answer key table. 
    It will take the location and ref/alt codons and combine them into one.
    It depends on find_modified_codons method and position_remover method."""
    input_tbl = input_tbl.move_to_start("Position_Ans")
    joined = input_tbl.with_column("Location+Codons", input_tbl.apply(location_and_codon_joiner_for_answer))
    joined = joined.move_to_start("Location+Codons")
    return joined
```

```python
ready_to_join_output_table = output_table_modifier(naglu_VEP_output)
ready_to_join_output_table.show()
```

```python
#review of answer key
naglu_answers
ready_to_join_answers = input_table_modifier(naglu_answers)
ready_to_join_answers.show()
```

```python
final_analysis_table = ready_to_join_answers.join("Location+Codons", ready_to_join_output_table)
final_analysis_table.to_csv("final_analysis_table.csv")
```
